```python
from nxviz.plots import CircosPlot
import networkx as nx
import os
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
import numpy as np
np.log10(times)[190]
```

```python
cluster_results = pkl.load(open('jaccard/markovstab_results.pkl','rb'))
times = cluster_results['times']
community_id = cluster_results['community_id']


graph = nx.read_gpickle(os.path.join("data","hox_gene_expression_jaccard.gpickle"))
#graph = nx.relabel_nodes(G, {i: "long name " + str(i) for i in range(len(G))})

i=70
for n in graph:
    graph.nodes[n]['pred2'] = int(community_id[i][n])
    
```

```python
c = CircosPlot(
    graph,
    node_grouping="neuron_class",
    node_color="pred2",
    #node_labels='neuron_class',
    figsize=(20, 20),
    group_order="alphabetically",
    group_label_position='middle',
    node_label_layout = 'rotation',
    edge_width="weight",
    node_label_rotation=True,
    fontsize=30
)

c.draw()
#plt.savefig('cluster_2.svg')
```
