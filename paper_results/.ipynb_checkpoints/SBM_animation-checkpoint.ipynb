```python
import numpy as np
import networkx as nx
import pylab as plt
from matplotlib.animation import FFMpegWriter
import copy

n = 50
graph = nx.stochastic_block_model([n, n], [[0.6, 0.02],
                                           [0.02, 0.6]], seed=1)

def random_disk(circle_x=0, circle_y=0, circle_r=1):
    # random angle
    alpha = 2 * np.pi * np.random.uniform()
    # random radius
    r = circle_r * np.sqrt(np.random.uniform())
    # calculating coordinates
    x = r * np.cos(alpha) + circle_x
    y = r * np.sin(alpha) + circle_y
    return np.array([x,y])

pos = {}
for i in range(n):
    pos[i] = random_disk(0,0)
    
for i in range(n,2*n):
    pos[i] = pos[i-n]
        
graph = nx.convert_node_labels_to_integers(graph)
```

```python
fig = plt.figure()
ax = plt.subplot(111)

t = np.arange(10)

pos_list = []
for i in t:
    pos1 = copy.deepcopy(pos)
    
    for j in range(n,2*n):
        pos1[j] = np.array([pos1[j][0] + i/10*3, pos1[j][1]])
        
    pos_list.append(pos1)

metadata = dict(title='Movie Test', artist='Matplotlib',comment='Movie support!')
writer = FFMpegWriter(fps=1, metadata=metadata)
with writer.saving(fig, "SBM.mp4", 100):
    for i in t:       

        if i!=0:
            edges.set_alpha(0)
            nodes.set_alpha(0)
            
        nodes = nx.draw_networkx_nodes(graph,pos_list[9-i],node_color='k',node_size=40,ax=ax)        
        edges = nx.draw_networkx_edges(graph,pos_list[9-i],width=1.0,alpha=0.5,ax=ax)       
        
        writer.grab_frame()
```
