```python
%load_ext autoreload
%autoreload 2
import numpy as np
import sys as sys
import os as os
import geocluster as gc
from geocluster import plotting
from graph_library import generate_barbell, generate_tree, generate_clique_of_cliques, assign_graph_metadata
import networkx as nx
import scipy as sc
import pylab as plt
import matplotlib.gridspec as gridspec
from matplotlib.animation import FFMpegWriter
from tqdm import tqdm
import itertools
```

```python
graph, pos = generate_clique_of_cliques(params = {'m':4, 'n': 3, 'L': 500, 'w':[1, 5, 10], 'p':[0.01, 0.1, 1]},
                                                  seed=0)
        
# initialise the code with parameters and graph 
graph.graph['name'] = 'clique_of_cliques'
graph = assign_graph_metadata(graph, pos=pos)

times = np.logspace(-4., 1., 30)
params = {'n_workers': 10, 'GPU': False, 'lambda': 0, 
          'with_weights': False, 'cutoff': 1.-1e-8, 'use_spectral_gap': True}

#Compute the OR curvatures
kappas = gc.compute_curvatures(graph, times, params, disable=True)

#Compute the zero-crossings of the edge curvatures
from scipy.interpolate import InterpolatedUnivariateSpline
zeros = []
for j in range(kappas.shape[1]):
    f = InterpolatedUnivariateSpline(np.log10(times), kappas[:,j], k=3)
    _zeros = f.roots()
    if len(_zeros)==0:
        zeros.append(-5) #these cross at -Infty, but set it to the lowest value
    if len(_zeros)>0:
        zeros.append(_zeros[0])
        
dic = {}
for i, e in enumerate(graph.edges):
    dic[e] = {'zeros': zeros[i]}

nx.set_edge_attributes(graph, dic)          
```

```python
fig = plotting.plot_graph(graph,zeros,node_size=30,edge_width=2,figsize=(10, 7),color_map=1,label='Zero-crossing')
```

```python
def update_graph(graph, threshold):
    
    graph_filt = graph.copy()
    
    #find edges (i,j) with zeros larger than threshold and adjacent nodes
    edgelist = [(i,j) for i, j, v in graph.edges(data=True) if v['zeros'] < threshold] 
#    nodelist = list(set([i for i, j, v in graph.edges(data=True) if v['zeros'] < threshold]))    
    
    #make new graph and find connected components
#    g_temp = nx.Graph()
#    g_temp.add_edges_from(edgelist)   
#    conn_comp = [c for c in sorted(nx.connected_components(g_temp), key=len, reverse=True)]
    
    #remove leaves (degree 1) nodes
    
#    for k, c in enumerate(conn_comp):
#        center = sum([graph_filt.nodes[i]['pos'] for i in list(c)])
#        graph_filt.add_node(num_nodes+k, pos=center)
        
        #connect all nodes adjacent to the connected component to the newly added node
#        for n in list(c):
#            nb = list(graph_filt.neighbors(n))
#            nb = list(itertools.product([num_nodes+k], nb))
#            graph_filt.add_edges_from(nb)
    num_nodes = nx.number_of_nodes(graph_filt)
    count = num_nodes
    while len(edgelist)>0: 
        (i,j) = edgelist[0]
        edgelist.remove((i,j))
    
        if i==j:
            continue
                 
        center = 0.5*(graph_filt.nodes[i]['pos'] + graph_filt.nodes[j]['pos'])
        graph_filt.add_node(count, pos=center)
        nbi = list(graph_filt.neighbors(i))
        nbi = list(itertools.product([count], nbi))
        nbj = list(graph_filt.neighbors(j))
        nbj = list(itertools.product([count], nbj))
        for e in list(graph_filt.edges(i)):     
            graph_filt.add_edge(count, e[1], zeros=graph_filt.edges[e]['zeros'])
            
        for e in list(graph_filt.edges(j)):    
            graph_filt.add_edge(count, e[1], zeros=graph_filt.edges[e]['zeros']) 
    
        graph_filt.remove_node(i)
        graph_filt.remove_node(j)   

        edgelist = [(e[0], count) if (e[0], i)==e or (e[0], j)==e else e for e in edgelist]
        edgelist = [(count, e[1]) if (i, e[1])==e or (j, e[1])==e else e for e in edgelist]
        edgelist = list(set(edgelist))
        
        count += 1
          
    print(len(graph_filt.nodes) )
    mapping = dict(zip(graph_filt.nodes(), range( num_nodes-int(len(edgelist)/2) )))
    graph_filt = nx.relabel_nodes(graph_filt, mapping) # nodes 1..26
    
    return graph_filt
    
```

```python
from ipywidgets import interact

%matplotlib inline
import matplotlib.pyplot as plt

def update_plot(threshold):
    graph_filt = update_graph(graph, threshold)
    roots_filt = [graph_filt.edges[i]['zeros'] for i in graph_filt.edges]
    fig = plotting.plot_graph(graph_filt.copy(),roots_filt,node_size=30,edge_width=2,figsize=(10, 7),color_map=1,label='Zero-crossing')
    plt.show()
    
zeros = nx.get_edge_attributes(graph,'zeros').values()     
zero_min = min(zeros)
zero_max = max(zeros)
interact(update_plot, threshold=(zero_min-0.1, zero_max+0.1))    
```
